```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#definiujemy funkcję do zminimalizowania
def f(x,y):
  return x**2+y**2

#definiujejmy różniczkę funkcji f(x,y) -> po x
def df_dx(x,y):
  return 2*x

#definiujejmy różniczkę funkcji f(x,y) -> po y
def df_dy(x,y):
  return 2*y
```

```python
#definicja spadku gradientu
def gradient_descent(start_x,start_y,learning_rate,num_iterations):
  x = start_x
  y = start_y

  history = []
  for i in range(num_iterations):
    grad_x = df_dx(x,y)
    grad_y = df_dy(x,y)

    x = x - learning_rate*grad_x
    y = y - learning_rate*grad_y

    history.append((x,y,f(x,y)))
  return x,y,f(x,y),history
```

```python
x_range = np.arange(-10,10,0.1)
y_range = np.arange(-10,10,0.1)
X,Y = np.meshgrid(x_range,y_range)
Z = f(X,Y)

start_x, start_y = 8,8
learning_rate = 0.1
num_iterations = 20
x_opt,y_opt,f_opt,history = gradient_descent(start_x,start_y,learning_rate,num_iterations)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z, cmap='coolwarm')
ax.scatter(*zip(*history), c='r', marker='o')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
```
